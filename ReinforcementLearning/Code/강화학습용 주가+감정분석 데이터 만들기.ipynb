```python
import pandas as pd
import numpy as np
```

```python
kospi20=['LG전자','이마트','기아차','SK하이닉스','한국전력',\
'현대백화점','셀트리온','KB금융','LG화학','아모레퍼시픽','기업은행','현대모비스','대우조선해양','GS건설',\
'S-Oil','삼성전자','현대차','대한항공','SK텔레콤'] #,'CJ제일제당'

for kospi in kospi20:
    v2_df = pd.read_csv("../../kospi_V2/data/{}_V2.csv".format(kospi))
    senti_df = pd.read_csv("../../Sentiment Analysis/Data/senti_{}.csv".format(kospi))

    del v2_df['Unnamed: 0.1']
    del v2_df['Unnamed: 0']

    rl_df = pd.merge(v2_df, senti_df, left_on=['Date'], right_on=['Date'], how='right')
    rl_df.to_csv("../../ReinforcementLearning/Data/{}_RL.csv".format(kospi), index=False)
```
