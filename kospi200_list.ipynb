```python
import time
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
driver = webdriver.Chrome('chromedriver.exe')
```

```python
url = 'https://namu.wiki/w/KOSPI200'
# 페이지 이동
driver.get(url)
time.sleep(1)
```

```python
#beautifulsoup으로 크롤링하기
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
soup
```

```python
table = soup.find("table", class_="wiki-table")
kospi200 = table.find_all("span",class_="wiki-color", style="color:#000")
```

```python
kospi200s = []
for i in range(len(kospi200)):
    kospi200s.append(kospi200[i].text)
```

```python
add = ['고려제강','한국단자','한세실업','대덕전자','한올바이오파마']
for i in add:
    kospi200s.append(i)
```

```python
print(kospi200s,len(kospi200s))
```

```python
kospi200_df = pd.DataFrame(kospi200s)
kospi200_df.write()
```
